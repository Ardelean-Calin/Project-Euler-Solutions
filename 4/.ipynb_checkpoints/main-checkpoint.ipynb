```python
def createPalindromeList():
    # creaza o lista cu toate palindroamele de 3 cifre
    palindromes = []
    for i in range(100, 1000):
        j = splitNumber(i)
        j.extend(reverseNumber(j))
        palindromes.append(createNumber(j))
    return palindromes
```

```python
def splitNumber(n):
    # imparte un numar in cifre
    digits = []
    while n > 0:
        digits.append(n%10)
        n = int(n/10)
    digits.reverse()
    return digits
```

```python
def createNumber(numberList):
    # creaza un numar intreg dintr-o lista de cifre
    nr = 0
    for item in numberList:
        nr *= 10
        nr += item
    return nr
```

```python
def reverseNumber(digitList):
    d = list(digitList)
    d.reverse()
    return d
```

```python
# Asa voi afla care e cel mai mare palindrom a carui 2 factori sunt mai mici de 1000

# creez lista ce contine toti palindromii de 6 caractere
palindromes = createPalindromeList()
# deoarece caut pe cel mai mare, incep de la cel mai mare si cobor
palindromes.reverse()
# pentru fiecare palindrom, vad daca se poate imparti cu un numar de 3 cifre
# astfel incat rezultatul sa fie tot un numar de 3 cifre. Dupa care iau cel
# mai mare numar, iar acela imi este rezultatul! :)
for item in palindromes:
    #flag = False
    for i in range(999, 100, -1):
        if item % i == 0 and int(item/i) < 1000:
            flag = True
            print(item, i)
            break
    #if flag: break
```
